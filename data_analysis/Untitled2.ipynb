```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('IPL_Squad_2023_Auction_Dataset.csv')
```

```python
records,feature=df.shape
print(f"Total_Records: {records}")
print(f"Total_feature: {feature}")
print("================================================")
print("================= compelete Information of over dataset ======================")
print(df.info)
print("================================================")
```

```python
print("Datatypes of Feature")
print(df.dtypes)
```

```python
print("Total Missing Value")
print(df.isnull().sum())
```

```python
print('=================== Sold_Player VS Unsold_Player(visualization) =====================')
unsold = (df['Team'] == 'Unsold').sum()
sold = (df['Team'] != 'Unsold').sum()
mylabels = ['Sold', 'Unsold']
plt.pie([sold,unsold],labels=mylabels)
plt.legend()
plt.show()
```

```python
print('======================  Clean_Data  ===========================')
print('Missing value after cleaning')
df = df[df["Team"] != "Unsold"].copy()
df['2022 Squad']=df['2022 Squad'].fillna("New_Player")
df.isnull().sum()
```

```python
print('============= Statistical_Summary =================')
df.describe()
```

```python
top10=df.nlargest(4, "COST IN ₹ (CR.)")
plt.bar(top10["Player's List"],top10["COST IN ₹ (CR.)"])

plt.xlabel("players")
plt.ylabel("COST IN ₹ (CR.)")
plt.ylabel("Price")
plt.title("Top 04 Most Expensive Players")
plt.show()
```
